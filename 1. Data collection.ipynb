# Data Retrieval:

Follow these steps to retrieve historical price data for Bitcoin, Ethereum, and Litecoin

### A. From Yahoo Finance

1) Go to the Yahoo Finance website (https://finance.yahoo.com)

2) In the search bar at the top of the page, type in the symbol for the cryptocurrency you want to retrieve data for. For example, for Bitcoin, type in "BTC-USD". For Ethereum, type in "ETH-USD". For Litecoin, type in "LTC-USD".

3) Press Enter to search for the cryptocurrency.

4) On the cryptocurrency's summary page, click on the "Historical Data" tab.

5) Select the time range you want to retrieve data for using the drop-down menu.

6) Choose the frequency you want the data to be displayed in (daily, weekly, or monthly).

7) Click on the "Download" button to download the data in CSV format.

```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
from datetime import datetime

```

```python
tickerSymbol = ['BTC-USD', 'ETH-USD', 'LTC-USD']
tickerData = yf.Tickers(tickerSymbol)
current_date = datetime.today().strftime('%Y-%m-%d')

tickerDf = tickerData.history(period='1d', start='2010-1-1', end=current_date)

# View the first 5 rows
print(tickerDf.tail())
```

```python
import coinmarketcapapi
client = coinmarketcapapi.CoinMarketCapAPI('YOUR_API_KEY')

response = client.cryptocurrency_listings_latest(limit=100)

# Print the symbol and price for each cryptocurrency
for currency in response.data:
    print(currency['symbol'], currency['quote']['USD']['price'])
```
